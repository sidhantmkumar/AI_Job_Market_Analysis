```python
def make_remote_bucket(val):
    try:
        s = str(val).strip().replace('%','')
        num = float(s)
    except:
        return 'Unknown'
    if num==0:
        return '0'
    if 1 <= num <= 49:
        return '1-49'
    if 50 <= num <= 99:
        return '50-99'
    if num==100:
        return '100'
    return 'Other'

df['remote_bucket'] = df['remote_ratio'].apply(make_remote_bucket)
df['remote_bucket'].value_counts()
```

```python
avg_by_bucket = df.groupby('remote_bucket')['salary_usd'].mean().reindex(['0','1-49','50-99','100','Other','Unknown'])
avg_by_bucket = avg_by_bucket.dropna()

plt.figure(figsize=(6,4))
plt.bar(avg_by_bucket.index, avg_by_bucket.values)
plt.title('Avg Salary by Remote Bucket')
plt.ylabel('Avg Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
from rapidfuzz import process, fuzz

# create a small canonical list
canonical = {
 'bachelors':'bachelors',
 "bachelor's":"bachelors",
 'b.sc':'bachelors',
 'bsc':'bachelors',
 'masters':'masters',
 "master's":"masters",
 'm.sc':'masters',
 'phd':'phd',
 'high school':'high_school'
}

# function to map value
def fuzzy_map_education(val):
    if pd.isna(val):
        return None
    val = str(val).lower().strip()
    choices = list(canonical.keys())
    match, score, _ = process.extractOne(val, choices, scorer=fuzz.token_sort_ratio)
    if score >= 75:
        return canonical[match]
    return val  # keep original if no good match

df['education_norm'] = df['education'].apply(fuzzy_map_education)
df['education_norm'].value_counts().to_csv('visuals/education_norm_counts.csv')
```
